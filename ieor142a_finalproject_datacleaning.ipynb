```python
# load the dataset
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_parquet("hf://datasets/lmsys/chatbot_arena_conversations/data/train-00000-of-00001-cced8514c7ed782a.parquet")
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.shape[0]
```

```python
# Step 3: Filter rows with empty converstaionts
df_clean = df.dropna(subset=["conversation_a", "conversation_b"])
df_clean.shape[0]
```

```python
print(df_clean['conversation_a'][0])
print(df_clean['conversation_b'][0])
print(type(df_clean['conversation_a']))
```

```python
df_clean['prompt'] = df_clean['conversation_a'].apply(lambda x: x[0]['content'])
df_clean['response_a'] = df_clean['conversation_a'].apply(lambda x: x[1]['content'])
df_clean['response_b'] = df_clean['conversation_b'].apply(lambda x: x[1]['content'])
df_clean.head()
```

```python
# Define a function to categorize questions into topics
def categorize_topic(prompt):
    if any(keyword in prompt.lower() for keyword in ['math', 'equation', 'number', 'algebra']):
        return 'Math'
    elif any(keyword in prompt.lower() for keyword in ['literature', 'poem', 'novel', 'author']):
        return 'Literature'
    elif any(keyword in prompt.lower() for keyword in ['science', 'physics', 'chemistry', 'biology']):
        return 'Science'
    elif any(keyword in prompt.lower() for keyword in ['history', 'war', 'ancient', 'civilization']):
        return 'History'
    elif any(keyword in prompt.lower() for keyword in ['technology', 'ai', 'machine learning', 'programming']):
        return 'Technology'
    else:
        return 'Other'

# Filter English questions and add a "topic" column
english_questions = df_clean[df_clean['language'] == 'English'].copy()
english_questions['topic'] = english_questions['prompt'].apply(categorize_topic)
english_questions['topic'].value_counts()
```

```python
# Sample 75 per major topic
topics_to_sample = ['Math', 'Literature', 'Science', 'History', 'Technology']
samples_list = []

for topic in topics_to_sample:
    topic_group = english_questions[english_questions['topic'] == topic]
    sampled = topic_group.sample(n=75, random_state=42)
    samples_list.append(sampled)

# Sample 200 from "Other" (because some topics are not properly classified)
other_group = english_questions[english_questions['topic'] == 'Other']
sampled_other = other_group.sample(n=200, random_state=42)
samples_list.append(sampled_other)

final_sampled_df = pd.concat(samples_list).reset_index(drop=True)
final_sampled_df.head()
```

```python
# Select relevant cols
final_sampled_df_labeling = final_sampled_df[[
    "prompt", "model_a", "model_b", "response_a", "response_b", "winner"
]].copy()
print("number of sampled rows: ", final_sampled_df_labeling.shape[0])
```

```python
# Add empty columns for manual labeling
final_sampled_df_labeling["Open/Closed"] = ""
final_sampled_df_labeling["Domain"] = ""
final_sampled_df_labeling["Comments"] = ""
final_sampled_df_labeling.head()
```

```python
# drop the column that presents a guideline violation
final_sampled_df_labeling = final_sampled_df_labeling[
    final_sampled_df_labeling['prompt'] != "YOUR INPUT VIOLATES OUR CONTENT MODERATION GUIDELINES. PLEASE TRY AGAIN."
].reset_index(drop=True)

# add another random column that from original dataset
# Find rows that are NOT in the sampled dataset

sampled_prompts = set(final_sampled_df_labeling['prompt'])
available_rows = df_clean[~df_clean['prompt'].isin(sampled_prompts)]

new_row = available_rows.sample(n=1, random_state=42)

# Format it
new_row_formatted = new_row[[
    "prompt", "model_a", "model_b", "response_a", "response_b", "winner"
]].copy()

new_row_formatted["Open/Closed"] = ""
new_row_formatted["Domain"] = ""
new_row_formatted["Comments"] = ""

# Add this new row to your existing dataframe
final_sampled_df_labeling = pd.concat([final_sampled_df_labeling, new_row_formatted], ignore_index=True)
print(final_sampled_df_labeling.shape[0])
```

```python
# Save
final_sampled_df_labeling.to_csv("chatbot_arena_sampled_topics.csv", index=False)

print(f"{len(final_sampled_df_labeling)} samples saved to 'chatbot_arena_sampled_topics.csv'.")
```
